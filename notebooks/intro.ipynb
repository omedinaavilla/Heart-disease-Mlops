# Proyecto predictivo de enfermedades cardiovasculares

Las enfermedades cardiovasculares son una de las principales causas de muerte a nivel mundial, según la OMS. Detectarlas a tiempo es fundamental para prevenir complicaciones graves. Con la digitalización de los historiales médicos y la disponibilidad de datos clínicos estructurados, es posible aplicar técnicas de aprendizaje automático para predecir el riesgo de falla cardíaca.

Se usara el dataset **“Heart Failure Prediction”** que contiene registros clínicos de pacientes, con variables como presión arterial, colesterol, glucosa, edad, entre otras para entrenar nuestros modelos.

El objetivo de este proyecto es desarrollar y comparar múltiples modelos de clasificación binaria para predecir la probabilidad de que un paciente presente riesgo de falla cardíaca (`HeartDisease = 1`) o no (`HeartDisease = 0`).

Para ello se implementarán y evaluarán distintos algoritmos de aprendizaje supervisado, entre los cuales se incluyen:
- <span style="color:#00BFFF; text-decoration: underline;">LogisticRegression</span>
- <span style="color:#00BFFF; text-decoration: underline;">RandomForestClassifier</span>
- <span style="color:#00BFFF; text-decoration: underline;">KNeighborsClassifier</span>
- <span style="color:#00BFFF; text-decoration: underline;">GradientBoostingClassifier</span>

El desarrollo se dividirá en tres etapas principales:
1. **Análisis exploratorio de datos (EDA)** para examinar la estructura y distribución de las variables.
2. **Preprocesamiento y detección de *data leakage*** mediante el uso de *pipelines* y validación cruzada.
3. **Entrenamiento, comparación y evaluación** de modelos con métricas como ROC-AUC, F1-score y accuracy.

Finalmente, el modelo con mejor rendimiento será empaquetado y desplegado mediante una **API**, permitiendo realizar predicciones automatizadas sobre nuevos pacientes en un entorno productivo.
